```python
import pandas as pd
import pycaret
```

```python
dados = pd.read_csv('jogos.csv')
dados.tail()
```

```python
dados['Data'] = pd.to_datetime(dados['Data'])

#FutureWarning: Series.dt.weekofyear and Series.dt.week have been deprecated. Please use Series.dt.isocalendar().week instead.
dados['Semana'] = dados['Data'].dt.week
dados.tail()
```

```python
#Now works without Warning
dados['Semana2'] = dados['Data'].dt.isocalendar().week
dados.tail()
```

```python
from pycaret.classification import *

#Before press ENTER: See "Semana2" is empty data type
s = setup(dados, target='Casa')
```
